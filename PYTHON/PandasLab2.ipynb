Sample data:

score={'Math':[78,85,96,80,86], 'English':[84,94,89,83,86],'Hindi':[86,97,96,72,83]}

1. add a new column for Science marks
2. also add another column total marks but get values from other columns to add marks into this column
3. add a new row to add the record of a new student

```python
score={'Math':[78,85,96,80,86], 'English':[84,94,89,83,86],'Hindi':[86,97,96,72,83]}
```

```python
import pandas as pd
df=pd.DataFrame(score)

df['Science']=[56,89,79,87,99]
df

# df['Total Marks']=[sum(df.columns)]
df['Total Marks']=df.sum(axis=1)

df
```

```python
newRow=pd.DataFrame([{"Math":45,"English":56,"Hindi":95,"Science":78}])
newRow["Total Marks"]=newRow.sum(axis=1)
df=pd.concat([df,newRow],ignore_index=True)
```

```python
data={"RollNo":[101,102,103,104],
      "sname":["Abhavay","Komal","Muskan","Shruti"],
      "course":["B.Tech-DS","B.Tech-AI","BCA","B.Sc"],
      "fees":[1000,1000,1000,1000]}
df=pd.DataFrame(data)
```

```python
# add a new column later -> city
df['city']=["Gurgaon","Noida","Delhi","Gurgaon"]
df
```

```python
pos = 3
df1 = df.iloc[:pos]
df2 = df.iloc[pos:]
newRow = pd.DataFrame([
    {"RollNo": 105, "sname": "Aradhya", "course": "B.Sc", "fees": 1000, "city": "Gurgaon"},
    {"RollNo": 106, "sname": "Niharika", "course": "B.tech-DS", "fees": 1000, "city": "Noida"}
])
df = pd.concat([df1, newRow, df2], ignore_index=True)
df
```

```python
# rename column headers to ->  rollno-> Roll_No, sname = Student_Name
df = df.rename(columns={"RollNo": "Roll_No", "sname": "Student_Name"})
df
```

```python
def renameRows(df):
    df.index = [f"row {i+1}" for i in range(len(df))]
    return df

df = renameRows(df)
df
```
