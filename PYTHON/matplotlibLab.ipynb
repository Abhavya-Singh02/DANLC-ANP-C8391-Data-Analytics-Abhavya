Lab1: Analyze the relationship between the size of houses (measured in square
footage) and their selling prices in a particular neighborhood. You have collected
data on various houses in that neighborhood.Create a scatter plot using the
below data and share your conclusion/analysis.
Input:
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800,
3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])


```python
import numpy as np
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
plt.scatter(square_footage, selling_prices, color="green")
plt.grid(True)
plt.title("Housing Price vs Square Footage")
plt.xlabel("Square Footage (sq, ft.)")
plt.ylabel("Selling price (in dollar $)")
plt.tight_layout()
plt.show()
```

Lab2: Create a pie chart to visualize the distribution of your monthly income by
source. You have collected data on the various sources of your income, such as
salary, freelance work, investments, and rental income. Share your
conclusion/analysis.
Input:
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]

```python
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]

plt.figure(figsize=(4,4))
plt.pie(monthly_income, labels=income_sources, shadow =True,autopct='%1.1f%%',startangle=140)
plt.title("Distribution of Monthly Income by Source")
plt.show()
```

Lab3: Suppose you're a sales manager for an e-commerce company, and you
want to create a figure with subplots to compare the sales performance of
different product categories over time. You have sales data for four product
categories: Electronics, Clothing, Home & Garden, and Sports & Outdoors. Share
your conclusion/analysis.


```python
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000,
36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000,
23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000,
25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000,
19000, 20000, 21000, 22000, 23000])

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(months, electronics_sales, 'o-')
axes[0, 0].set_title("Electronics")
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Sales Amount')

axes[0, 1].plot(months, clothing_sales, 'o-g')
axes[0, 1].set_title("Clothing")
axes[0, 1].set_xlabel('Month')
axes[0, 1].set_ylabel('Sales Amount')

axes[1, 0].plot(months, home_garden_sales, 'o-r')
axes[1, 0].set_title("Home & Garden")
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Sales Amount')

axes[1, 1].plot(months, sports_outdoors_sales, 'o-b')
axes[1, 1].set_title("Sports & Outdoors")
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Sales Amount')

plt.tight_layout()
plt.show()
```

Scenario:Analyzing Sales Data
Suppose you work for a retail company, and you have dummy data containing sales
data for the past year. The data includes information such as SalesDate,product
names,regions, sales quantities, prices, and dates. You have to generate a bar chart
,pie plot on region and prices and line chart on SalesDate and prices columns.
Further, you need to get some inference out of the chart.
Create a ChatGPT prompt to generate the code for this scenario. Based on the code
generated, ask ChatGPT to give the conclusion/inference.
Note. You can provide the data to ChatGPT or ask it to use sample data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating dummy data
data = {
    'SalesDate': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'ProductName': ['Product A'] * 3 + ['Product B'] * 3 + ['Product C'] * 3 + ['Product D'] * 3,
    'Region': ['North', 'South', 'East', 'West'] * 3,
    'SalesQuantity': [150, 120, 130, 160, 140, 155, 180, 170, 165, 175, 160, 150],
    'Price': [250, 270, 260, 255, 240, 245, 250, 265, 275, 280, 290, 300]
}

df = pd.DataFrame(data)

# Bar Chart: Total Sales Quantities by Region
region_sales = df.groupby('Region')['SalesQuantity'].sum()
plt.figure(figsize=(8, 6))
region_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales Quantities by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales Quantity')
plt.show()

# Pie Chart: Proportion of Sales Revenue by Region
df['Revenue'] = df['SalesQuantity'] * df['Price']
region_revenue = df.groupby('Region')['Revenue'].sum()
plt.figure(figsize=(8, 6))
region_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'red'])
plt.title('Proportion of Sales Revenue by Region')
plt.ylabel('')  # Hide y-label for a cleaner pie chart
plt.show()

# Line Chart: Price Trend Over Time
plt.figure(figsize=(10, 6))
plt.plot(df['SalesDate'], df['Price'], marker='o', color='purple')
plt.title('Price Trend Over Time')
plt.xlabel('Sales Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
